```python
import pandas as pd
from itertools import product, combinations
from pathlib import Path

# è®€å– Excel
file_path = "../fault.xlsx"
df = pd.read_excel(file_path, dtype=str)

display(df)
```

```python
# æ‰€æœ‰ 4-bit ç‹€æ…‹
all_states = [''.join(bits) for bits in product('01', repeat=4)]

# åˆ¤æ–·æŸç‹€æ…‹æ˜¯å¦èƒ½è¦†è“‹æŸ fault
def is_covered(state, fault):
    for i, f in enumerate([fault['Da'], fault['Dv'], fault['Ia'], fault['Iv']]):
        if pd.isna(f):
            continue
        if state[i] != f:
            return False
    return True

# å»ºç«‹ state -> å¯è¦†è“‹çš„ fault åç¨±åˆ—è¡¨
state_cover_map = {s: [] for s in all_states}
for idx, row in df.iterrows():
    for state in all_states:
        if is_covered(state, row):
            state_cover_map[state].append(row['Name'])
```

```python
# æ‰¾å‡ºèƒ½å®Œæ•´è¦†è“‹æ‰€æœ‰ faults çš„æœ€å°ç‹€æ…‹å­é›†
def find_exact_minimal_state_cover(state_cover_map, all_faults):
    state_list = list(state_cover_map.keys())
    for r in range(1, len(state_list) + 1):
        for subset in combinations(state_list, r):
            covered = set()
            for s in subset:
                covered.update(state_cover_map[s])
            if covered >= all_faults:
                return list(subset)
    return []

all_faults = set(df['Name'])
exact_states = find_exact_minimal_state_cover(state_cover_map, all_faults)
```

```python
out_path = Path("../fault_with_f.xlsx")   # è¼¸å‡ºæª”æ¡ˆ

for s in exact_states:
    df[s] = df['Name'].apply(lambda n: 1 if n in state_cover_map[s] else pd.NA).astype('Int64')

base_cols = ['Name', 'SFR', 'Da', 'Dv', 'Ia', 'Iv', 'Ops', 'Detect', 'F', 'R', 'Gv']
final_cols = base_cols + exact_states
df[final_cols].to_excel(out_path, index=False)

print("âœ… å®Œæˆï¼Œæª”æ¡ˆè¼¸å‡ºè‡³:", out_path)
```
